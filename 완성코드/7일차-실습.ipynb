### 막대(bar) 그래프
- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.text.html?highlight=pyplot%20text#matplotlib.pyplot.text

```python
import matplotlib.pyplot as plt

# 한글, 그래프 라벨 마이너스 표시, 그래프 크기
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['figure.figsize'] = [10, 5]  # (6, 4)

# 년도별 3.0이상 지진강도 횟수
X = list(range(2005,2015))
Y = [15,7,2,10,8,5,14,9,18,8]

plt.bar(X, Y)     # 세로 막대 그래프
#plt.barh(X, Y)   # 가로 막대 그래프
plt.title('년도별 그래프', fontsize=15)
plt.xlabel('년도')
plt.ylabel('지진강도')
for x, y in zip(X, Y):     
    plt.text(x, y, str(y), fontsize=12, fontweight='bold', c='r', 
             horizontalalignment='center', 
             verticalalignment='top')  # x,y좌표에 글자 넣기

plt.show()
```

#### - 막대별로 색깔 다르게 표현하기

```python
colors = ['b','g','r','c','m','y','k']

plt.bar(X, Y, color=colors)     # 세로 막대 그래프
#plt.barh(X, Y)   # 가로 막대 그래프
plt.title('년도별 그래프', fontsize=15)
plt.xlabel('년도')
plt.ylabel('지진강도')
for x, y in zip(X, Y):     
    plt.text(x, y, str(y), fontsize=15, fontweight='bold', c='w', 
             horizontalalignment='center', 
             verticalalignment='top')  # x,y좌표에 글자 넣기

plt.grid()
plt.show()
```

##### 2개의 막대 그래프로 나타내기

```python
# 년도별 3.0이상 지진강도 횟수
import numpy as np

X = list(range(2005,2015))
label = X
print(X)
Y1= [15,7,2,10,8,5,14,9,18,8]
Y2= [y+1 for y in Y1]
X = np.arange(len(X))
print(X)

plt.bar(X-0.2, Y1, width=0.4, label='Y1', edgecolor='k')
plt.bar(X+0.2, Y2, width=0.4, label='Y2', edgecolor='k') 

plt.title('년도별 그래프', fontsize=15)
plt.xlabel('년도')
plt.ylabel('지진강도')
plt.xticks(X, label)
plt.grid()
plt.legend()

plt.show()
```

#### 혼합(막대 + 선) 그래프

```python
import numpy as np

X = list(range(2005,2015))
Y1= [15,7,2,10,8,5,14,9,18,8]  # 선
Y2= [y+1 for y in Y1]          # 막대


plt.plot(X, Y1, '^', linestyle='dashed', c='r', 
         linewidth=2, label='Y1')
for x, y in zip(X, Y1):     
    plt.text(x, y, str(y), fontsize=15, fontweight='bold', c='w', 
             horizontalalignment='center', 
             verticalalignment='center')

plt.bar(X, Y2, label='Y2')
for x, y in zip(X, Y2):     
    plt.text(x, y, str(y), fontsize=15, fontweight='bold', c='k', 
             horizontalalignment='center', 
             verticalalignment='center')
plt.legend()
plt.title('년도별 그래프', fontsize=15)
plt.xlabel('년도')
plt.ylabel('지진강도')
plt.show()
```

```python
plt.pie(val, labels=idx, autopct='%.1f%%',
        textprops={'color':'b','size':15})   # 텍스트색상이 흰색이 아니도록 주의!!!
plt.title('학년별 동아리 인원수', fontsize=15)
plt.show()
```

#### - legend 위치 지정
- bbox_to_anchor=(x, y, width, height) 

```python
plt.pie(val, labels=idx, autopct='%.1f%%',
        textprops={'color':'b','size':15})
plt.legend(idx, title='학년', 
           loc='center right', 
          bbox_to_anchor=(1, 0, 0.5 ,1))
plt.title('학년별 동아리 인원수', fontsize=15)
plt.show()
```

#### - 파이 조각이 분리되도록 나타내기

```python
explodes = [0.2, 0.05, 0.05, 0.05, 0.05, 0.05] # 전체 돌출(비율별로)
explodes = [0.1, 0, 0, 0, 0, 0]   # 특정 조각만 돌출

plt.pie(val, labels=idx, autopct='%.1f%%',
        textprops={'color':'b','size':15},
       explode=explodes)
plt.legend(idx, title='학년', 
           loc='center right', 
          bbox_to_anchor=(1, 0, 0.5 ,1))
plt.title('학년별 동아리 인원수', fontsize=15)
plt.show()
```

#### - 그림자(shadow) 표시하기

```python
explodes = [0.1, 0, 0, 0, 0, 0]   # 특정 조각만 돌출

plt.pie(val, labels=idx, autopct='%.1f%%',
        textprops={'color':'b','size':15},
       explode=explodes, shadow=True)
plt.legend(idx, title='학년', 
           loc='center right', 
          bbox_to_anchor=(1, 0, 0.5 ,1))
plt.title('학년별 동아리 인원수', fontsize=15)
plt.show()
```

#### - 각도(startangle) 조정

```python
explodes = [0.1, 0, 0, 0, 0, 0]   # 특정 조각만 돌출

plt.pie(val, labels=idx, autopct='%.1f%%',
       textprops={'color':'b','size':15},
       explode=explodes, shadow=True,
       startangle=40)
plt.legend(idx, title='학년', 
           loc='center right', 
          bbox_to_anchor=(1, 0, 0.5 ,1))
plt.title('학년별 동아리 인원수', fontsize=15)
plt.show()
```

#### - 그래프의 열을 병합하여 나타내기

```python
import matplotlib.pyplot as plt
import numpy as np

grid = plt.GridSpec(2, 3)
#grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.3)

# 1행 선그래프
X = np.linspace(0,10,100) 
Y = np.cos(X)
plt.subplot(grid[0, :2]).plot(X, Y)

# 1행 이미지그래프
Z = np.random.uniform(0,1,(5,5))  # 0~1사이의 균일분포값을 5x5형태로
plt.subplot(grid[0, 2]).imshow(Z)

# 2행 점그래프
X = np.random.randn(100)  # 정규 분포 데이터
Y = np.random.randn(100)  # 정규 분포 데이터
plt.subplot(grid[1, 0]).scatter(X, Y)

# 2행 막대그래프
X = np.arange(10)
Y = np.random.uniform(1,10,10) # 1~10 사이의 균일분포
plt.subplot(grid[1, 1:]).bar(X, Y)
plt.show()
```

### [실습] 행, 열 병합하여 그래프 그리기

```python
import matplotlib.pyplot as plt
import numpy as np

grid = plt.GridSpec(3, 3, wspace=0.4, hspace=0.3)

# 1행 점그래프
X = np.random.randn(100)
Y = np.random.randn(100)
plt.subplot(grid[0, :]).scatter(X, Y)
# 2행 선그래프
X = np.linspace(0,10,100)
Y = X*2
plt.subplot(grid[1, :2]).plot(X, Y)
# 2-3행 이미지그래프
Z = np.random.uniform(0,1,(10,5))
plt.subplot(grid[1:, 2]).imshow(Z)
# 3행 선그래프
X = np.linspace(0,10,100)
Y = np.cos(X)
plt.subplot(grid[2, 0]).plot(X, Y)
#3행 막대그래프
X = np.arange(10)
Y = np.random.uniform(1,10,10)
plt.subplot(grid[2, 1]).bar(X, Y)

plt.show()
```

### 히스토그램
- 주어진 데이터를 몇개의 구간으로 나누어서 각 구간의 (빈)도수와 분포를 분석하는 그래프

```python
heights = np.array([175,165,164,164,171,165,160,169,164,159,
                    163,167,163,172,159,160,156,162,166,162,
                    158,167,160,161,156,172,168,165,165,177])
plt.hist(heights, bins=6, edgecolor='w')
plt.title('히스토그램(frequency) 그래프')
plt.xlabel('height')
plt.ylabel('도수')
plt.show()
```
