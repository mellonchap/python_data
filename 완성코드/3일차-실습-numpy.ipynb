### [실습] 1차원 배열의 벡터화 연산

```python
import numpy as np

x = np.arange(1, 21)

print(x)
print(f'1.3의 배수: { x[x % 3 == 0] }')
print(f'2.4로 나누어 1이 남는 수 : {x[x % 4 == 1] }')
print(f'3.3으로 나누어지고 4로 나누면 1이 남는 수: {x[(x % 3 == 1) & (x % 4 == 1)]}')
```

### [실습] 넘파이 벡터화 연산을 주식 구매 대금 구하기
- @ : 벡터연산 결과 내적 구하기

```python
p = np.array([100,80,50]) # 주식금액
n = np.array([3,4,5])     # 주식수량

total = p * n
print(f'주식구매대금: {np.sum(total)}만원')

total = p @ n  # 벡터연산 결과 내적 구하기
print(f'주식구매대금: {total}만원')
```

### 4.기술통계 함수

numpy는 아래와 같은 데이터 집합에 대해 간단한 통계를 계산하는 함수(통계량)를 제공한다.
 - 데이터의 개수(count): **len()**
 - 평균(mean, average): **numpy.mean()** 평균을 통계용어->표본 평균(sample average, sample mean)
 - 분산(variance): **numpy.var()** 표본 분산, 데이터와 표본 평균간의 거리의 제곱의 평균, 표본 분산이 작으면 데이터가 모여있는 것이고 크면 흩어져 있는 것
 - 표준 편차(standard deviation) :**numpy.std()**  표본 분산의 양의 제곱근 값
 - 최댓값(maximum) : **numpy.max()** 데이터 중에서 가장 큰 값
 - 최솟값(minimum) : **numpy.min()** 데이터 중에서 가장 작은 값
 - 최대값의 인덱스(argmax): **numpy.argmax()** 
 - 최소값의 인덱스(argmin): **numpy.argmin()** 
 - 중앙값(median) : **numpy.median()** 데이터를 크기대로 정렬하였을 때 가장 가운데에 있는 수
 - 사분위수(quartile) : **numpy.percentile(x, 25)** 데이터를 가장 작은 수부터 가장 큰 수까지 크기가 커지는 순서대로 정렬하였을 때 1/4, 2/4, 3/4 위치에 있는 수

```python
x = np.array([18,5,10,23,19,-8,10,0,0,5,2,15,8,2,5,4,15,-1,4,-7,-24,7,9,-6,23,-13])
print(x)
```

#### 위 함수에 적용해 보기

#### 히스토그램 그리기

```python
x = np.array([18,5,10,23,19,-8,10,0,0,5,2,15,8,2,5,4,15,-1,4,-7,-24,7,9,-6,23,-13])
print(len(x))

hist, bins = np.histogram(x, bins=5)
print(f'구간: {bins}')
print(f'도수: {hist}')


import matplotlib.pyplot as plt

#plt.hist(x, bins, edgecolor='w')
#plt.hist(x, bins=5, edgecolor='w')
hist, bins, patches = plt.hist(x, bins=5, edgecolor='w')
print(f'구간: {bins}')
print(f'도수: {hist}')
print(f'오브젝트: {patches}')

plt.show()

```

### [실습] 주사위 100번 던져서
- 각 숫자 개수 세기
- 평균 구하기

```python
x = np.random.randint(1, 7, size=100)
np.sort(x)
print(f'x: {x}' )

print(f'요소들의 개수: {np.bincount(x)[1:]} ')
print(f'평균: {np.mean(x)}')
```

```python
import matplotlib.pyplot as plt

plt.hist(x, edgecolor='w') # 전처리 안하고 데이터를 넣어도 된다.
plt.show()
```

```python
plt.bar(list(range(1,7)), np.bincount(x)[1:])
plt.show()
```
