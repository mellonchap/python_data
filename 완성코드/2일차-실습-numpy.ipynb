### [실습] 이미지(흑백: 2차원)를 1차원 벡터로 변경하기

```python
from sklearn.datasets import load_digits 
import matplotlib.pyplot as plt

digits = load_digits()

print(digits.images.shape)
# 이미지 샘플데이터를 리스트에 저장
samples = [0, 10, 20, 30, 1, 11, 21, 31] # 선택된 이미지번호
d = []  # 이미지 저장 리스트
for i in range(8):
    d.append(digits.images[samples[i]])
    

# 8개 이미지를 하나의 그래프에 나타내기
plt.figure(figsize=(8, 2))  # 그래프 크기: inch
for i in range(8):
    plt.subplot(1, 8, i+1)
    plt.imshow(d[i], interpolation='nearest', cmap=plt.cm.bone_r)
    plt.title(f'image:{i+1}')
plt.suptitle('Number: 0 & 1 image')
plt.tight_layout()
plt.show()
```

#### 1차원 벡터로 변형

```python
# 이미지를 1차원 벡터화하기
v = []
for i in range(8):
    v.append(d[i].reshape(64,1))  # 벡터화

# 1차원 벡터 그래프 나타내기    
plt.figure(figsize=(8, 2))  # 그래프 크기: inch
for i in range(8):
    plt.subplot(1, 8, i+1)
    plt.imshow(v[i], aspect=0.3, interpolation='nearest', cmap=plt.cm.bone_r)
    plt.title(f'Vector:{i+1}')
plt.suptitle('Vector: 0 & 1 image', y=1.05)
plt.tight_layout(w_pad=7)
plt.show()
```

### 배열의 연결

#### - hstack
- 행의 수가 같은 두 개 이상의 배열 가로로 연결

```python
a = np.zeros((2, 2), dtype=int)
b = np.ones((2, 4), dtype=int)
print(a)
print(b)
print('-----------')
print( np.hstack([a, b]) )
```

#### - vstack
- 열의 수가 같은 두 개 이상의 배열 세로로 연결

```python
a = np.zeros((4, 4), dtype=int)
b = np.ones((2, 4), dtype=int)
print( np.vstack([a, b]) )
```

#### - dstack 
- 행이나 열이 아닌 깊이(면: depth) 방향으로 배열을 합침, 가장 안쪽 원소의 차원 증가 
- 3차원 데이터

#### # 1차원 데이터일 경우

```python
import numpy as np
na1 = np.zeros(2, dtype=int)    # [0 0]
na2 = np.ones(2, dtype=int)     # [1 1]
na3 = np.full(2, 2,  dtype=int) # [2 2]
na = np.dstack([na1, na2])
print(na.shape)
print(na)
```

```python
na = np.dstack([na1, na2, na3])
print(na.shape)
print(na)
```

#### #2차원 데이터일 경우

```python
na1 = np.zeros((2,3), dtype=int)   # [[0 0 0]
                                   #  [0 0 0]]
na2 = np.ones((2,3), dtype=int) 
na3 = np.full((2,3), 2,  dtype=int)

na = np.dstack([na1, na2])
print(na.shape)
print(na)
```

```python
na = np.dstack([na1, na2, na3])
print(na.shape)
print(na)
```

#### - stack : axis인수를 사용하여 사용자가 지정한 차원(축으로) 배열을 연결
- default axis=0 
- 1차원 : (행,  )      : (axis=0, ) 
- 2차원 : (행, 열)     : (axis=0, axis=1) 
- 3차원 : (행, 열, 면) : (axis=0, axis=1, axis=2)
- 배열의 크기가 같아야함
- 차원 증가

#### # 1차원 데이터

```python
na1 = np.zeros(2, dtype=int)    # [0 0]
na2 = np.ones(2, dtype=int)     # [1 1]
na3 = np.full(2, 2,  dtype=int) # [2 2]

na = np.stack([na1, na2])  # default: axis=0 행 쌓기
print(na)
print(na.shape)
```

```python
na = np.stack([na1, na2], axis=1) # 열로 쌓기
print(na)
print(na.shape)
```

```python
na = np.stack([na1, na2, na3])
print(na.shape)
print(na)
```

```python
na = np.stack([na1, na2, na3], axis=1)
print(na.shape)
print(na)
```

#### # 2차원 데이터

```python
na1 = np.zeros((2,3), dtype=int)   # [[0 0 0]
                                   #  [0 0 0]]
na2 = np.ones((2,3), dtype=int) 
na3 = np.full((2,3), 2,  dtype=int)

na = np.stack([na1, na2])
print(na.shape)
print(na)
```

```python
na = np.stack([na1, na2], axis=1)
print(na.shape)
print(na)
```

```python
na = np.stack([na1, na2, na3], axis=0)
print(na)
print(na.shape)
```

```python
na = np.stack([na1, na2, na3], axis=1)
print(na)
print(na.shape)
```
