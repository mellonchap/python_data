- $f(x) = x^3 -3x^2 + x $

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**3 -3 * x**2 + x

x = np.linspace(-1, 3, 9)
y = f(x)

print(f'x     : {x}')
print(f'y=f(x): {y}')      

plt.plot(x, y, 'o-r')
plt.axhline(y=0, color='black', linewidth=1)
plt.axvline(x=0, color='black', linewidth=1)
plt.title('Function f(x)=$x^3-3x^2+x$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()

plt.fill_between(x[2:], y[2:], alpha=0.5 )

plt.show()
```

#### 7. 다변수함수 (Multivariate function)
- 복수의 입력변수를 갖는 함수: z = f(x, y)
- 3차원 서피스 플롯(surface plot)/컨투어 플롯(contour plot)으로 나타냄
- 3차원 자료를 시각화 하는 방법: 명함이 아닌 등고선(contour)을 사용함
- $f(x,y)=2x^2+6xy+7y^2-26x-54y+107$

```python
def f(x, y):
    return 2*x**2 + 6*x*y + 7*y**2 -26*x -54*y + 107

x = np.linspace(-3, 7, 100)
y = np.linspace(-3, 7, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)


fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(X, Y, Z, linewidth=0.1)
ax.view_init(40, -110)
plt.xlabel('$x$')
plt.ylabel('$y$')
ax.set_zlabel('$z$')
plt.title('Surface Plot')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (10, 4)
CS = plt.contour(X, Y, Z, levels=np.logspace(0,3,10))
plt.clabel(CS,  fmt='%d')
plt.title('컨투어 플롯(Contour plot)')
plt.show()
```

#### 8. 분리 가능 다변수 함수
- 다변수함수의 곱으로 표현 가능한 다변수 함수
- f(x, y) = f1(x)f2(y)
- 2차원 함수에 대한 1차원 단변수함수는 3차원 공간상의 지형을 케이크 자르듯이 나눌 때 생가는 단변 모양과 같다. 그리고 분리 가능 다변수함수는 단면 모양이 모두 같다.
- $g(x, y) = exp(-x^2-16y^2) = exp(-x^2)exp(-16y^2) = g1(x)g2(y)$

```python
from matplotlib import transforms
from matplotlib.ticker import NullFormatter

def g1(x):
    return np.exp(-x**2)

def g2(y):
    return np.exp(-16*y**2)

def g(x, y):
    return g1(x) * g2(y)

xx = np.linspace(-1, 1, 100)
yy = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(xx, yy)
Z = g(X, Y)

left, width = 0.1, 0.65
bottom, height = 0.1, 0.65
bottom_h = left_h = left + width + 0.02
rect  = [left, bottom, width, height]
rectx = [left, bottom_h, width, 0.2]
recty = [left_h, bottom, 0.2, height]

plt.figure(1, figsize=(8,8))
ax = plt.axes(rect)
axx = plt.axes(rectx)
axy = plt.axes(recty)

nullfmt = NullFormatter()
axx.xaxis.set_major_formatter(nullfmt)
axy.yaxis.set_major_formatter(nullfmt)

ax.contour(X, Y, Z)
ax.axhline(0, c='r', ls='-')
ax.axhline(0.1, c='g', ls='--')
ax.axhline(0.2, c='b', ls='-.')
ax.axhline(0.3, c='c', ls=':')
ax.set_xlabel('x')
ax.set_ylabel('y')

axx.plot(xx, g1(xx), c='r')
axx.plot(xx, g2(0.1)*g1(xx), c='g', ls='--')
axx.plot(xx, g2(0.2)*g1(xx), c='b', ls='-.')
axx.plot(xx, g2(0.3)*g1(xx), c='c', ls=':')         
axx.set_title('g1(x)')
axx.text(-0.2, 0.3, 'y=0.3로 자른 단면')
axx.text(-0.2, 0.6, 'y=0.2로 자른 단면')
axx.text(-0.2, 0.9, 'y=0.1로 자른 단면')

base = axy.transData
rot = transforms.Affine2D().rotate_deg(-90)
axy.plot(yy, g2(yy), transform=rot + base)
axy.set_title('g2(y)')
axy.axhline(0, xmax=g2(0), c='r', ls='-')
plt.text(0.05, 0.02, '$g_2$(0)={:.2f}'.format(g2(0)))
axy.axhline(0.1, xmax=g2(0.1), c='g', ls='--')
plt.text(0.05, 0.12, '$g_2$(0.1)={:.2f}'.format(g2(0.1)))
axy.axhline(0.2, xmax=g2(0.2), c='b', ls='-.')
plt.text(0.05, 0.22, '$g_2$(0.2)={:.2f}'.format(g2(0.2)))
axy.axhline(0.3, xmax=g2(0.3), c='c', ls=':')
plt.text(0.05, 0.32, '$g_2$(0.3)={:.2f}'.format(g2(0.3)))

axx.set_xlim(ax.get_xlim())
axx.set_ylim(0, 1)
axy.set_xlim(ax.get_ylim())
axy.set_ylim(0, 1)

plt.suptitle('$g(x,y) = \exp(-x^2)\exp(-16y^2) = g_1(x)g_2(y)$의 그래프', y=1.04)
plt.show()
```

#### 9. 다변수 다출력 함수(소프트맥스 함수)
- 입력변수뿐 아니라 출력변수도 여러 개 일 경우 출력을 벡터나 행렬로 나타낼 수 있다.
- 소프트맥수(Softmax) 함수는 다차원 벡터를 입력 받아서 다차원 벡터로 출력한다.
- 출력 벡터는 다음과 같은 특성이 있다.
> - 모든 출력 원소는 0, 1 사이값을 갖는다.
> - 모든 출력 원소의 합은 1이다.
> - 입력 원소의 크기 순서와 출력 원소의 크기 순서가 같다.
- 소프트맥스 함수는 다변수 입력을 확률처럼 보이게 출력한다.
> - 이러한 특성 대문에 인공신경망의 마지막 단계에서 출력을 조건부확률로 변경하는데 사용함

```python
def softmax(x, w):  # x:1차원 행렬, w:가중치 벡터
    e = np.exp( w * x)
    return np.exp(w * x) / e.sum()

x = [2.0, 1.0, 0.5]
y = softmax(x, np.ones(3))

print(f'x : {x}')
print(f'y : {y}')
print(f'sum(y): {np.sum(y)}')

plt.plot(x, y)
plt.title('Softmax function')
plt.show()
```

### 미분법

#### 데이터 분석에서 미분이 필요한 이유

> [예제] 저절나사의 각도 x를 설정하여 출력되는 음량 y를 제엉하는 앰프가 있을 때 
> ---> 출력 음량(y)이 최대가 되는 조절나사의 각도(x)를 찾는 문제 Y = WX + b
>  미분이 사용됨

- **데이터 분석 목표**: 예측 오차가 가장 적은 최적의 예측 모형을 구하는 것
- 예측 모형의 계수(모수)를 바꾸어 최적의 예측 모형을 찾는다. 이때 성능을 비교할 수 있어야한다.
- 모수를 입력 받아서 오차(오류의 크기)를 출력하는 함수를 가장 작게 하는 것--> 이런 함수를 목적함수라고 한다.
> - 손실함수(loss fi=unction),  비용함수(cost function), 오차함수(error function)
- 최적화: 목적함수가 주어졌을 때 이 목적함수의 값을 가장 크게 또는 가장 적게 할 수 있는 합려값(모수)를 구하는 것
- **미분**: 입력값이 변했을 대 함수의 출력값이 어떻게 달라지는지를 정량적으로 알아야 하는데 이 과정이 미분이다.

##### 기울기
- **변화율(y증감분/x증감분)**에 따른 기울기(slope) 혹은 민감도(sensitivity)를 찾는 것
- 수치적 최적화: 가장 적은 횟수로 여러 x값을 시도하여 최적의 값을 찾는 방법

##### 3차 함수의 그래프의 기울기를 그래프로 그려보기

$x^3 -3x^2 + x$

- 접점에서의 기울기 = 접점에서의 미분계수 ($3x^2 - 6x + 1$)
- if x==0일 때의 기울기 ---> 빨간 점에서의 기울기
- if x==1일 때의 기울기 ---> 녹색점에서의 기울기 

```python
def f(x):
    return x**3 -3 * x**2 + x

x = np.linspace(-1, 3, 400)
y = f(x)

plt.plot(x, y)
plt.plot(0, 0, 'ro')
plt.plot(x, x, 'r:')
plt.plot(1, -1, 'go')
plt.plot(x, (3*1**2-6*1+1)*(x-1)-1, 'g--')

plt.xlim(-3.5, 5.5)
plt.ylim(-4, 2)
plt.xticks(np.arange(-3, 6))
plt.yticks(np.arange(-4, 2))

plt.annotate('', xy=(1,0), xytext=(0,0), arrowprops=dict(facecolor='gray'))
plt.annotate('', xy=(1,1), xytext=(1,0), arrowprops=dict(facecolor='gray'))

plt.annotate('', xy=(2,-1), xytext=(1,-1), arrowprops=dict(facecolor='gray'))
plt.annotate('', xy=(2,-3), xytext=(2,-1), arrowprops=dict(facecolor='gray'))

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('함수의 기울기')
plt.grid()
plt.show()
```

### [실습] 고객 만족도 점도표로 나타내기

```python
import matplotlib.pyplot as plt 
import numpy as np

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['figure.figsize'] = (10, 4)

# X축 Y축 데이터
index = ['매우만족','만족','보통','불만족','매우불만족']
data  = [5,11,21,9,4]

# 점도표를 위해 meshgrid()를 이용해,X,Y데이터를 가로 세로의 평면 배치로 만든다.
X = np.arange(len(index)) +1  # X축: index
Y = np.arange(1, max(data)+1)  # Y축: data(도수)
x, y = np.meshgrid(X, Y)       # x,y 평면 범위(격자형태)

# 점도표 그리기: 
# Y축이 실제값보다 작을 때까지 찍기
#plt.scatter(x, y, c= y<=data, cmap="Greys") # c=The marker colors:array-like or list of colors 
plt.scatter(x, y, c= y<=data, cmap="Oranges")
plt.xlabel('만족도')
plt.xticks(ticks=X, labels=index)
plt.title('표 3-1: 고객 만족도', size=15)
plt.show()
```

### [실습] 동아리 회원수를 비율그래프(Pie)로 나타내기



```python
import matplotlib.pyplot as plt 
import pandas as pd

# 1.표 만들기
datas= {1:16 ,2:12 ,3:7, 4:5}
df = pd.DataFrame(datas, index=['인원'])
df.columns.name= '학년'

# 2.행과열 바꾸기
df.columns = ['1학년','2학년','3학년','4학년']
df2= df.T
df2.index.name ='학년'
df2.columns = ['도수']

# 3.컬럼 추가하기
df2['상대도수']  = [x/df2.도수.sum() for x in df2.도수]
df2['백분율(%)'] = [x/df2.도수.sum()*100 for x in df2.도수]


#4. Pie그래프로 그리기
val = df2.도수
idx = df2.index
explodes = [0.1, 0, 0, 0]   # 특정 조각만 돌출

plt.pie(val, labels=idx, autopct='%.1f%%',
        textprops={'color':'black','size':15},
       explode=explodes, shadow=True)
plt.legend(idx, title='학년', 
           loc='center right', 
          bbox_to_anchor=(1, 0, 0.5 ,1))
plt.title('학년별 동아리 인원수', fontsize=15)
plt.show()
```

### [실습] 양적자료 점도표로 나타내기

```python
import matplotlib.pyplot as plt 
import numpy as np

datas = [4.1, 4.2, 3.8, 4.2, 3.9, 4.2, 3.9, 4.1, 3.9, 4.3,
         3.9, 3.8, 3.8, 4.0, 4.3, 3.8, 3.9, 4.1, 4.1, 4.0]

#1.고유한 측정값 찾기
index = list(set(datas))     # np.unique(datas)
index.sort()    # 정렬하기
data = [ datas.count(i) for i in index ]
print(f'index: {index}')
print(f'data: {data}')


#2.계급구간 만들기
X = index
Y = np.arange(1, max(data)+1)  # Y축:도수
x, y = np.meshgrid(X, Y)         # x-y 평면 범위(격자형태)

#3.점도표 그리기
plt.figure(figsize=(6, 2)) # 그래프 사이즈
plt.scatter(x, y, c= y<=data, cmap="Greys") # data는 배열데이터
plt.show()
```
