### [실습] 예제 1-10 : 경우의 수 구하기
- 주사위를 두 번 던져서 나온 눈의 수의 합이 소수인 경우

```python
import sympy
dice1 = [1,2,3,4,5,6]  # list(range(1,7))
dice2 = [1,2,3,4,5,6]
case_list = []

for i in dice1 :
    for j in dice2:
        if sympy.isprime(i + j) :
            case_list.append((i, j))
print(f'경우의수: {len(case_list)}, 사건: {case_list}')
```

```python
import sympy

dice1 = list(range(1,7))
dice2 = list(range(1,7))

d = [(x, y) for x in dice1 for y in dice2 if sympy.isprime(x+y) ]

print(f'경우의수: {len(d)}')
print(f'사건: {d}')
```

### [실습] 예제 1-11: 서로다른 두 점 연결하는 선분

```python
x = ['A', 'B', 'C', 'D', 'E']
y = ['A', 'B', 'C', 'D', 'E']
line = []

for i in x:
    for k in y:
        if i != k and k + i not in line:
            line.append(i + k)
print(line)
```

#### 데이터 선택의 경우의 수를 나열하는 방법
- itertools 라이브러리에서
- 반복(X) + 순서(O) : 순열 permutations() 
- 반복(X) + 순서(X) : 조합 combinations()
- 반복(O) + 순서(O) : 중복순열 product() --> sample
- 반복(O) + 순서(X) : 중복조합 combinations_with_replacement() --> selection

### [실습] 순열: 4개의 문자를 서로 다르게 나열하는 방법
- 반복(X) + 순서(O) : 순열 permutations()

```python
import itertools

S = {'A','B','C','D'}
r = 4   # 길이=선택 개수
N = list( itertools.permutations('ABCD', 4) )
print(f'순열의 경우의 수: {len(N)}')
print(N)
```

### [실습] 조합: 4개의 문자를 서로 다르게 나열하는 방법

```python
import itertools

S = {'A','B','C','D'}
r = 2   # 길이=선택 개수
N = list( itertools.combinations('ABCD', r) )
print(f'순열의 경우의 수: {len(N)}')
print(N)
```
