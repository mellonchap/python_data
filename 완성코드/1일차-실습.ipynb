#### 부분집합

```python
A = {2, 4, 6, 8, 10}
B = {1, 2, 3, 4, 5}
C = {2, 4, 8}
D = set()  # 공집합

print(B.issubset(A))   # B ⊂ A
print(C.issubset(A))   # C ⊂ A
print(D.issubset(A))   # D ⊂ A , 공집합도 부분집합
print(A.issubset(A))   # A ⊂ A , 자기 자신도 부분집합
```

### [실습] 예제 1-2 집합 구하기
- (a) $ A ∩ B $ 
- (b) $ A ∪ B $ 
- (c) $ A - B $ 
- (d) $ A ∩ B^c $ 

```python
U = set(range(1, 11))
A = {1, 3, 5, 7}
B = {3, 4, 5, 6}

#1. A ∩ B
print(f'A ∩ B: {A & B} ')
#2. A ∪ B
print(f'A ∪ B: {A | B}')
#3. A - B
print(f'A - B: {A - B}')
#4. A ∩ B^c
print(f'A ∩ B^c: {A & (U ^ B)}')
```

```python
import matplotlib.pyplot as plt

a, b, c = 2, 1, 2 
X = list(range(-10, 31))
Y = [ (a*x**2 + b*x + c) for x in X ] 

plt.plot(X, Y, linestyle='solid', c='r')
plt.grid()
plt.title('$y=2x$')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```
