#### - 주소의 일부값(도/시) 변경하기 : 신규 컬럼(주소2) 추가

##### 방법1: 
(주소정보 리스트화하여  --> 주소 변경--> 리스트를 DataFrame--> DataFrame.concat() 함수 적용)

```python
# 표준 행정구역 시/도 정보로 변환하기
addr = []   # df에서 주소만 뽑아서 스페이스로 분리된 주소 정보 리스트
for address in df['주소']:
    addr.append(str(address).split())
addr 

# 참고할 행정구역 정보
addr_dict={'서울':'서울특별시', '서울시':'서울특별시', 
           '부산시':'부산광역시','부산':'부산광역시', 
           '인천':'인천광역시', '인천시':'인천광역시',
           '대전시':'대전광역시', 
           '대구시':'대구광역시', 
           '울산시':'울산광역시',
           '전주시':'전주광역시',
           '광주':'광주광역시',   
           '경기':'경기도',
           '경북':'경상북도',
           '경남':'경상남도',
           '전남':'전라남도',
           '제주도':'제주특별자치도', '제주시':'제주특별자치도', 
           '충북':'충청북도', 
           '충남':'충청남도',
           '강원':'강원도'} 


addr2= []
for i in range(len(addr)):
    if addr[i][0] in addr_dict.keys():
        addr[i][0] = addr_dict[addr[i][0]]
        addr2.append(' '.join(addr[i]))
    else:
        addr2.append(' '.join(addr[i]))
#addr2

# 표준화된 주소 정보를 '주소2' 컬럼으로 만들어 기존 데이터와 합친다.
addr2 = pd.DataFrame(addr2, columns=['주소2'])
df = pd.concat([df, addr2], axis=1)  # 열 추가하기

df.head()
```

## 2. 네이버 OpenAPI이용해서 책 검색 데이터 분석(웹) 

### [실습] 네이버 OpenAPI이용해서 책 검색 데이터 추출하여 DataFrame으로 출력하기

```python
# 네이버 검색 API예제는 블로그를 비롯 전문자료까지 호출방법이 동일하므로 blog검색만 대표로 예제를 올렸습니다.
# 네이버 검색 Open API 예제 - 블로그 검색
import os
import sys
import urllib.request

import json
import pandas as pd

# 네이버 책검색 API ID/PW
client_id = "WlIpA7IkqcaBRP6XKyUk"  # 자신의 client id 
client_secret = "pIew6Mp_Fr"        # 자신의 client secret

# 가져올 데이터 개수
count = 100

def request(title):
    # 검색 URL 설정
    encText = urllib.parse.quote(title)
    url = "https://openapi.naver.com/v1/search/book.json"
    query = "?query=" + encText  # Jason 결과
    option = f"&display={count}&sort=count"   

    # 웹에 검색 URL로 요청하기
    request = urllib.request.Request(url + query + option )
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    response = urllib.request.urlopen(request)
    rescode = response.getcode()

    if(rescode==200):
        response_body = response.read()
        #print(response_body.decode('utf-8'))
        result = response_body.decode('utf-8')
    else:
        print("Error Code:" + rescode)

    # 검색 결과를 json type으로 간편하게 보여주기
    j_result = json.loads(result)
    if j_result == None:
        print('json.loads 실패!!!')
        exit()
        
    return j_result


def get_dicdata(j_result):
    # items > title, author, link
    titles, authors, links = [], [], [] 

    for idx, item in enumerate(j_result['items']):
        #print( item['title'], '---', item['author'])
        tmp = item['title'].replace('<b>','')
        tmp = tmp.replace('</b>','')
        titles.append(tmp)
        authors.append(item['author'])
        links.append(item['link'])

    # DataFrame으로 표현하기    
    books = {'제목':titles, '저자':authors, 'url':links}
    
    return books
    

    
#-----------------------
# 검색어 키보드로 입력 받기
#-----------------------
title = input('책 검색 단어 입력: ')
print('---------------------')

j_result = request(title)
books = get_dicdata(j_result)

df = pd.DataFrame(books)


# DataFrame을 CSV파일로 저장하기
file = 'data/search_book_result.csv'
df.to_csv(file, encoding='cp949')
        
# CSV파일 읽어서 DataFrame으로 표현하기
df = pd.read_csv(file, encoding='cp949')
df
```

## 3. 네이버 검색(뉴스) 이용해서 뉴스 데이터 분석(정적 웹 크롤링) 

### [실습] 네이버 검색(뉴스) 이용해서 뉴스 데이터 제목&URL 추출하여 DataFrame으로 출력하기

```python
from bs4 import BeautifulSoup
import requests

# 검색어 입력
search = input('검색할 키워드 입력')
search = search.replace(' ', '+')

url = 'https://search.naver.com/search.naver?where=news&query={}&start=11'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) Chrome/100.0.48496.75'}

#html불러오기
original_html = requests.get(url.format(search), headers=headers)
#original_html = requests.get(url)
html = BeautifulSoup(original_html.text, 'html.parser')

# 검색결과
articles = html.select('ul.list_news > li > div > div > a')

#뉴스기사 제목 가져오기
new_titles, urls = [], []
for i in articles:
    new_titles.append(i.attrs['title'])
    urls.append(i.attrs['href'])
    
titles = {'제목':new_titles, 'url':urls}

import pandas as pd
df = pd.DataFrame(titles)
df


```
