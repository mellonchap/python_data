### 실습문제1. 업다운 게임 만들기

(1~100)사이의 비밀의 숫자를 생성시킨 후 키보드로 입력받아 숫자를 맞추는 업다운 게임을 만들어 보세요. 아래와 같이 출력되도록 입력 시도 횟수도 함께 출력해 주세요.

```python
import random

number = random.randint(1,100)   #1~100의 랜덤한 숫자 생성
# print('-'*50)
# print('* 비밀의 숫자가 지정되었습니다.')
# print('-'*50)
ment = '''---------------------
* 비밀의 숫자가 지정되었습니다.
---------------------'''
print(ment)

trycnt = 0
while True:
    guess = int(input('숫자를 입력하세요 : '))
    trycnt += 1   # trycnt = trycnt + 1
    if guess == number:
        print('정답입니다')
        break
    elif guess > number:
        print('비밀의 수는 더 작은 수 입니다')
    else:
        print('비밀의 수는 더 큰 수 입니다.')

print('-'*50)
print(f'비밀의 숫자: {number}')
print(f'당신의 입력 횟수: {trycnt}')
```

### 실습문제2. 커피 자동주문 머신 만들기

아래 조건을 만족하는 커피 자동주문 머신을 만들어보세요.<br><br>
[커피 자동주문 머신]<br>
친구들과 현금 커피 자동주문 머신 앞에서 
 - 아메리카노 2잔과 카페라테 2잔 카푸치노 1잔을 주문하였다. 
 - 아메리카노 한잔에 2500원이고
 - 카페라테는 한잔에 3000원
 - 카푸치노는 한잔에 3000원이다. 
커피 종류와 잔 수를 결정한 후 현금 15000원을 투입하고 잔돈을 1000원 받았다. 

```python
c1, c2, c3 = 2500, 3000, 3000

def coffeeMachine(coffee1, coffee2, coffee3):
    hap = coffee1*c1 + coffee2*c2 + coffee3*c3
    return hap


menu = '''[커피 자동주문 머신 메뉴]
------------------------------
-아메리카노는 2500원
-카레라떼는 3000원
-카푸치노는 3000원
입니다. 원하시는 커피종류와 잔수를 입력하세요.
------------------------------\n'''
coffee1 = int(input(menu+"아메리카노 몇잔? : "))
coffee2 = int(input("카페라떼 몇잔? : "))
coffee3 = int(input("카푸치노 몇잔? : "))

hap = coffeeMachine(coffee1, coffee2, coffee3)

print('------------------------------')
print("지불할 총 금액은 ", hap, "원 입니다")
print('------------------------------')
money = int(input("돈을 넣어주세요~: "))
print('------------------------------')

change = money - hap
if change < 0:
    print("입력한 금액이 부족합니다. ", change*-1, "원을 넣어주세요.")
else:
    print("거스름돈은 ", change, "원 입니다")
    
```

### 실습문제3. 영어단어장 만들기

아래 조건을 만족시키는 <영어단어장 만들기> 프로그램을 만들어 보세요.

영단어는 key, 그의 뜻은 value 로 두는 dictionary 를 만드세요.<br>
<메뉴>
 - 1. 단어 등록
 - 2. 단어 검색
 - 3. 단어 삭제
 - 4. 모든 단어 보기
 - 0. 종료하기

1. 단어 등록 <br>
    새 영단어와 그의 뜻을 입력 받고 dictionary 에 추가합니다.

2. 단어 검색<br>
    영단어를 입력 받아 그의 뜻을 출력합니다.<br>
    미등록된 단어인 경우 '미등록 단어'를 출력합니다.
    
3. 단어 삭제<br>
    영단어를 입력 받아 단어장에서 해당 단어를 삭제한다.<br>
    미등록된 단어인 경우 '미등록 단어'를 출력합니다
    
4. 모든 단어 보기<br>
    현재 등록되어 있는 모든 단어와 뜻을 출력합니다.

0. 종료하기

```python
menu = '''---선택메뉴--
1. 단어 등록
2. 단어 검색
3. 단어 삭제
4. 모든 단어 보기
0. 종료하기
------------------------------'''

# 빈 딕셔너리를 만든다. (반복문 안에 넣지 않는다!)
wordbook = {}

def addBook():
    w = input('새 단어 입력: ')
    m = input(f'{w}의 뜻 입력: ')
    wordbook[w] = m
    
def searchBook():
    w = input('검색할 단어 입력: ')
    if w in wordbook:
        print(f'{w} : {wordbook[w]}')
    else:
        print(f'미등록 단어입니다.')
        
def deleteBook():
    w = input('삭제할 단어 입력: ')
    if w in wordbook:
        del wordbook[w]
        print(f'{w} 단어 삭제되었습니다!')
    else:
        print(f'미등록 단어입니다.')
        
def showAllBook():
    print('현재 딕셔너리에 있는 단어 목록입니다.')
    print('-'*20)
    for w, m in wordbook.items():
        print(f'{w} : {m}')
    
    
while True:
    select = input(menu+'선택:')

    if select == '0':
        print('프로그램을 종료합니다.')
        break
    elif select == '1':          #1. 단어 등록
        addBook()        
    elif select == '2':          #2. 단어 검색
        searchBook()
    elif select == '3':          #3. 단어 삭제
        deleteBook()
    elif select == '4':          #4. 모든 단어 보기
        showAllBook()
    else:
        print('잘못 입력하셨습니다.')
        
print('-'*30)
print('영어 단어장 프로그램이 종료되었습니다.')
```

### 실습문제4. 온라인 장바구니 클래스 만들기

온라인 장바구니를 구현하는 클래스를 만들어 보세요. 

```python
# 온라인 장바구니 구현
class Basket:
    
    def __init__(self,id):
        self.id = id       #인스턴스ID
        self.items = []    #장바구니 물품명
        self.prices = []   #장바구니 물품가격
        self.quantity = [] #장바구니 물픔수량
        self.total = 0     #장바구니 전체가격
        self.itemcnt = 0   #장바구니 전체수량
        
    #장바구니에 추가    
    def add(self, item, price, qty):
        self.items.append(item) 
        self.prices.append(price)
        self.quantity.append(qty)
        self.total += price * qty
        self.itemcnt += 1

    #장바구니에서 삭제
    def delete(self, item, qty):
        for i in range(self.itemcnt):
            if item == self.items[i]:
                self.quantity[i] -= qty
                self.total -= self.prices[i] * qty
                if self.quantity[i] == 0:
                    self.itemcnt -=1
                    del self.items[i]
                    del self.quantity[i]
                    del self.prices[i]
                break

    #장바구니에 물건 출력
    def printitems(self):
        print(f'#{self.id}의 장바구니\n--------------')
        for i in range(self.itemcnt):
            print('*',self.items[i], self.prices[i], self.quantity[i])
        print("** 전체가격 =", self.total, ", 전체품목 =", self.itemcnt)
        print('-'*30)

basket1 = Basket("제인")
basket2 = Basket("대원")

basket1.add("바나나", 3000,2)
basket1.add("우유", 2500, 2)
basket1.add("라면", 4000, 1)
basket2.add("라면", 4000, 2)
basket1.delete("우유", 1)

basket1.printitems()
basket2.printitems()
```

```python
# 온라인 장바구니 구현
class Basket:
    def __init__(self, id):
        self.id = id
        self.바구니={}
        
    def add(self, item, price, qty):
        self.바구니[item]=[price,qty]
        
    def delete(self, item, qty):
        #수량이 < 0일 경우에 대한 처리를 해주세요.
        self.바구니[item][1] -= qty
        
    def printitems(self):
        print(f'# {self.id}의 장바구니')
        print('-'*30)
        total = 0
        cnt = len(self.바구니)
        
        for key in self.바구니:
            print(f'* {key:10} {self.바구니[key][0]} {self.바구니[key][1]}')
            total += self.바구니[key][0] * self.바구니[key][1]
        print(f'** 전체가격 =  {total} 전체 품목 = {cnt}')
        
basket1 = Basket("제인")
basket2 = Basket("대원")

basket1.add('바나나',3000,2)
basket1.add('우유',2500,2)
basket1.add('라면',4000,1)
basket2.add('라면',4000, 2)

basket1.delete('우유',1)

basket1.printitems()
basket2.printitems()
```
